<b><font color='red'>Ejercicio 01</font></b>  

Representa la función $\frac{D}{(x^2+1)}$ , donde D es el día de tu cumpleaños  
Hazque el título sea tenga una fuente: serif de color blue y tamaño 16

<img src="f1.png" ALIGN="left">


```python
import matplotlib.pyplot as plt
import numpy as np
D=31
x = np.linspace(-5,5,500)
y = D / (x**2 + 1)

plt.plot(x,y,color="green",label=r'$\frac{31}{x^2 + 1}$')
font={'family':'serif', 'color':'blue','size':16}
plt.title("Alan Mikolajczyk", fontdict=font)
plt.grid(True)
plt.legend()
plt.show()
```

<b><font color='red'>Ejercicio 02</font></b>  

Representa también la función con D/2, busca y representa también como poner las barras verticales y horizontales

<img src="f2.png" ALIGN="left">

```python
import matplotlib.pyplot as plt
import numpy as np
D=31
d=31/2
x = np.linspace(-5,5,500)
y1= D / (x**2 + 1)
y2= d / (x**2 + 1)

plt.plot(x,y1,color="green",label=r'$\frac{31}{(x^2 + 1)}$')
plt.plot(x,y2,linestyle='--',color='red',label=r'$\frac{15.5}{(x^2 + 1)}$')

x_line = 0
y_line = d / (x_line**2 + 1)

plt.axvline(x=2, color='cyan', linestyle='--')
plt.axvline(x=-2, color='cyan', linestyle='--')
plt.plot([2, -2], [y_line, y_line], color='cyan', linestyle='--')

font={'family':'serif', 'color':'b','size':16}
plt.title("Alan Mikolajczyk", fontdict=font)
plt.grid(True)
plt.legend()
plt.show()
```
